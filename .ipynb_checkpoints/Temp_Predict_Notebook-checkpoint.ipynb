# 1. Which country has the most severe overall temperature change from decade to decade? Please elaborate on how you interpret severe change and decade to decade.

```python
df_cou['year'] = pd.DatetimeIndex(df_cou['dt']).year
df_cou['decade'] = df_cou['year'].astype(str).str.slice(0, -1)+'0' # Replace last digit of year with 0 to get decade. 
```

```python
df_cou['scaled_avg_temp'] = (df_cou['AverageTemperature']-df_cou['AverageTemperature'].min())/(df_cou['AverageTemperature'].max()-df_cou['AverageTemperature'].min())+0.000001
df_cou['log_avg_temp'] = np.log(df_cou['scaled_avg_temp'])
```

The reason we normalize the avg temperatures and then scale them into their logarithms is that for example a temperature change from 1 to 2 degrees would constitute a 100% change, but so would a change from 20 to 40 degrees. Since a change from 1 to 2 degrees obviously occurs more frequently than one from 20 to 40 degrees between two decades, leaving the data 'untreated' would result in a heavy bias towards colder countries and the list with most severe changes would have almost only cold countries at the top.

```python
list_country = list(df_cou['Country'].unique())
```

```python
mean_temp = {} # create a dictionary to store the results
for country in list_country:
    a = df_cou.loc[df_cou['Country'] == country] # Create a dataframe for each country
    b = a.groupby(['decade'])['log_avg_temp'].mean() # get the mean temperature for each decade
    c = list(b.pct_change()*100) # get the percentual change in mean temperature from decade to the next decade, which is my definition of severe change
    d = [abs(x) for x in c] # change can be positive or negative so we are only interested in absolute change
    e = np.nanmean(d) # get the mean percentage of temperature change of a country for all decades - basically one value per country
    mean_temp[country] = e # add country and mean percentage as key/value pair to out dictionary
```

```python
for key, value in mean_temp.items():
    mean_temp['Antarctica'] = 0 # all avg temperature values for antarctica are nan's, so we r
        
mean_temp_sorted = {k: v for k, v in sorted(mean_temp.items(), key=lambda item: item[1], reverse=True)}
list(mean_temp_sorted.items())[:5]
```

# 2. Now write a function which returns N entries of cities which have the highest variability​ in AverageTemperature in a specified time range.

```python
df_cit = pd.read_csv('GlobalLandTemperaturesByCity.csv')
```

```python
def variability(df, 
                dt_col, 
                city_col, 
                cnt_col,
                long_col,
                avg_temp_col, 
                start_date, 
                end_date, 
                n_cities=4):
    if start_date < end_date:
        df_new = df.copy()
        df_new['city_country'] = df_new[city_col]+' // '+df_new[cnt_col]+' // ' + df_new[long_col]
        df_new[dt_col] = pd.to_datetime(df_new[dt_col], format='%Y-%m-%d')
        df_new = df_new[(df_new[dt_col] >= start_date) & (df_new[dt_col] <= end_date)]
        df_grp = df_new.groupby(['city_country'])[avg_temp_col].std().sort_values(ascending=False)[:n_cities]
        cities = df_grp.index.to_list()
        print(df_grp)
        return df_new[df_new['city_country'].isin(cities)]

    else:
        raise ValueError('Start Date be more recent than End Date')
```

```python
df_var = variability(df_cit, 'dt', 'City', 'Country','Longitude','AverageTemperature', '2005-01-01', '2007-01-01')
```

```python
heihe_train_normalized = torch.FloatTensor(heihe_train_normalized).view(-1)
```

```python
train_window = 70
```

The in_out_seqsfunction will accept the raw input data and will return a list of tuples. In each tuple, the first element will contain list of 70 items corresponding to the sum of revenues in 70 months, the second tuple element will contain one item - the sum of revenue on the 70th+1st month.

```python
def in_out_seqs(input_data, tw):
    inout_seq = []
    L = len(input_data)
    for i in range(L-tw):
        train_seq = input_data[i:i+tw]
        train_label = input_data[i+tw:i+tw+1]
        inout_seq.append((train_seq ,train_label))
    return inout_seq
```

```python
train_inout_seq = in_out_seqs(heihe_train_normalized, train_window)
```


input_size: Corresponds to the number of features in the input. Though our sequence length is 70, for each day we have only 1 value - the avg temperature, therefore the input size will be 1.

hidden_layer_size: Specifies the number of hidden layers along with the number of neurons in each layer. We will have one layer of 100 neurons.

output_size: The number of items in the output, since we want to predict the avg temperature for 1 month in the future, the output size will be 1.

in the constructor we create variables hidden_layer_size, lstm, linear, and hidden_cell. LSTM algorithm accepts three inputs: previous hidden state, previous cell state and current input. The hidden_cell variable contains the previous hidden and cell state. The lstm and linear layer variables are used to create the LSTM and linear layers.

Inside the forward method, the input_seq is passed as a parameter, which is first passed through the lstm layer. The output of the lstm layer is the hidden and cell states at current time step, along with the output. The output from the lstm layer is passed to the linear layer. The predicted temperature is stored in the last item of the predictions list, which is returned to the calling function.

```python
class LSTM(nn.Module):
    def __init__(self, input_size=1, hidden_layer_size=100, output_size=1):
        super().__init__()
        self.hidden_layer_size = hidden_layer_size

        self.lstm = nn.LSTM(input_size, hidden_layer_size)

        self.linear = nn.Linear(hidden_layer_size, output_size)

        self.hidden_cell = (torch.zeros(1,1,self.hidden_layer_size),
                            torch.zeros(1,1,self.hidden_layer_size))

    def forward(self, input_seq):
        lstm_out, self.hidden_cell = self.lstm(input_seq.view(len(input_seq) ,1, -1), self.hidden_cell)
        predictions = self.linear(lstm_out.view(len(input_seq), -1))
        return predictions[-1]
```

The next step is to create an object of the LSTM() class, define a loss function and the optimizer. Since, we are solving a regression problem, we will use the mean squared error loss. For the optimizer function, we will use the adam optimizer.

```python
model = LSTM()
loss_function = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```


Model training is 100 epochs, with the loss printed out every 20th epoch. As you can see, probably more epochs and some parameter tuning is needed for better convergence, but for this show

```python
epochs = 100

for i in range(epochs):
    for seq, labels in train_inout_seq:
        optimizer.zero_grad()
        model.hidden_cell = (torch.zeros(1, 1, model.hidden_layer_size),
                        torch.zeros(1, 1, model.hidden_layer_size))

        y_pred = model(seq)

        single_loss = loss_function(y_pred, labels)
        single_loss.backward()
        optimizer.step()

    if i%20 == 1:
        print(f'epoch: {i:3} loss: {single_loss.item():10.8f}')

print(f'epoch: {i:3} loss: {single_loss.item():10.10f}')
```

the prediction lags behind the true value because the series is autoregressive (i.e. a strong way to predict next months’s value is “It will be about the same as this month”). The model therefore corrects itself with the new information when it misses a big jump. In other words, if the temperature jumps one month and the model does not predict that, it has learned to take into account the higher temperature for the prediction of the next month’s temperature.
